```python
#functions
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from nltk import tokenize
import spacy
#öznelerden biri from/to daysa direkt herşey doğru, özne yoksa doğru, öznelerin hiçbiri hiçbi ülkeye uymuyosa doğru,
#öznelerden başka ülkeye ait olan varsa (from, todan yoksa) sadece to'yu o ülkeyle değiş. 
nlp = spacy.load("en_core_web_sm")

kw = [["CA", "Canada", "Ottawa", "Justin Trudeau", "Trudeau", "Canadian", "Canadians"],
["CN","China", "Beijing", "Xi Jinping", "Jinping", "Chinese" ],
["DE", "Germany", "Berlin", "Angela Merkel", "Merkel", "German", "Germans"],
["FR", "France", "Paris", "French", "Emmanuel Macron", "Jean Castex", "Macron", "Castex","Édouard Philippe", "Philippe"],
["IR", "Iran", "Tehran", "Iranian", "Hassan Rouhani", "Rouhani", "Ali Khamenei", "Khamenei"],
["IT", "Italy", "Rome", "Italians", "Sergio Mattarella", "Mattarella", "Mario Draghi", "Draghi", "Giuseppe Conte", "Conte"],
["JP", "Japan", "Tokyo", "Japanese", "Yoshihide Suga", "Suga", "Shinzo Abe", "Abe", "Fumio Kishida", "Kishida"],
["RU", "Russia", "Moscow", "Vladimir Putin", "Putin", "Dmitry Medvedev", "Medvedev", "Mikhail Mishustin", "Mishustin"],
["UK", "United Kingdom", "Kingdom",  "U.K.", "Great Britain", "GB", "G.B.","Britain", "British", "London", "Boris Johnson", "Johnson"],
["US", "U.S.", "United States", "States", "DC", "D.C.", "Washington", "America", "Americans", "Joe Biden", "Biden", "Donald Trump","Trump", "Mike Pence","Pence","Kamala Harris","Harris"]]

def country_index(country_code):
    if country_code == 'CA':
        return 0
    elif country_code == 'CN':
        return 1
    elif country_code == 'DE':
        return 2
    elif country_code == 'FR':
        return 3
    elif country_code == 'IR':
        return 4
    elif country_code =='IT':
        return 5
    elif country_code == 'JP':
        return 6
    elif country_code == 'RU':
        return 7
    elif country_code == 'UK':
        return 8
    elif country_code == 'US':
        return 9
    else:
        return -1
    

def find_subject(header, fro, to):
    from_index = country_index(fro)
    to_index = country_index(to)
    sent = header
    doc=nlp(sent)
    subject_tokens = [tok for tok in doc if (tok.dep_ == "nsubj") ]
    for sbj_token in subject_tokens:
        
        if (sbj_token.text in kw[from_index]) or (sbj_token.text in kw[to_index]):
            return to
    for sbj_token in subject_tokens:
        for list_count, each_list in enumerate(kw):      #iterates through each list in lst
            if sbj_token.text in each_list:                    #checks if specified character is in each_list
                     
                return kw[list_count][0]
            
    return to



    
def vader_score_compound(paragraph):              ## for tokenizing text
    
    sentence_list = tokenize.sent_tokenize(paragraph)
    paragraphSentiments = 0.0
    for sentence in sentence_list:
        sentiment_dict = analyzer.polarity_scores(sentence)
    
        paragraphSentiments += sentiment_dict['compound']
    
    return round(paragraphSentiments / len(sentence_list), 4)

def vader_score_compound_sentence(sentence):          #for raw 
    sentiment_dict = analyzer.polarity_scores(sentence)
    return sentiment_dict['compound']

def vader_score_positive(sentence):
    sentiment_dict = analyzer.polarity_scores(sentence)
    return sentiment_dict['pos']

def vader_score_neutral(sentence):
    sentiment_dict = analyzer.polarity_scores(sentence)
    return sentiment_dict['neu']

def vader_score_negative(sentence):
    sentiment_dict = analyzer.polarity_scores(sentence)
    return sentiment_dict['neg']

def find_kw(subject):
    for list_count, each_list in enumerate(kw):      #iterates through each list in lst
        if subject in each_list:                    #checks if specified character is in each_list
            return spec_char, kw[list_count][0]  #returns both the character and list that contains the character
```

```python
# find scores with vader

do_tokenize = True 
look_for_subjects = True

df['fulltext'] = df['header'] + df['body'] 

if do_tokenize:
    df['compound'] = df['fulltext'].apply(vader_score_compound)
else:
    df['compound'] = df['fulltext'].apply(vader_score_compound_sentence)
    
if look_for_subjects:    
    df['to_analyzed'] = df.apply(lambda x: find_subject(x['header'], x['from'], x['to']), axis=1)
    
    df_changed = df[(df.to != df.to_analyzed)] #degisenleri asagida gosterme amaciyla 
    display(df_changed) 
    
    df['to'] = df['to_analyzed']

display(df)
```

```python
# find average
import numpy as np
array1 = df[["from", "to", "date", "compound"]].to_numpy()

print('Scores for each news:\n')
print(array1)

df['avg'] = df.groupby(['from','to'])['compound'].transform('mean')
df2= df.drop_duplicates(subset=['from', 'to'])
#display(df2)


array2 = df2[["from", "to", "avg"]].to_numpy()
print('\nAverage scores matrix for each country pair:\n')
print(array2)

df3 = df2.pivot(index='from', columns='to', values='avg')
df3= df3.fillna(0)

display(df3)
```
